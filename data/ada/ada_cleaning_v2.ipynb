# EDA for Average Daily Attendance Data

***Import libraries:***

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

***Data Reading:***

```python
# read in the csv file
df_d = pd.read_csv('ada_district.csv')

print(df_d.info())
df_d.head()
```

```python
cols_old = df_d.columns.to_list()
cols_new = ['County #',
'County Name',
'District #',
'District Name',
'2004-2005 ADA',
'2005-2006 ADA',
'2006-2007 ADA',
'2007-2008 ADA',
'2008-2009 ADA',
'2009-2010 ADA',
'2010-2011 ADA',
'2011-2012 ADA',
'2012-2013 ADA',
'2013-2014 ADA',
'2014-2015 ADA',
'2015-2016 ADA',
'2016-2017 ADA',
'2017-2018 ADA',
'2018-2019 ADA',
'2019-2020 ADA',
'2020-2021 ADA']

df_d.rename(columns=dict(zip(cols_old, cols_new)), inplace=True)
```

```python
df_d.replace(',', '', regex=True, inplace=True)
df_d[cols_new[4:]] = df_d[cols_new[4:]].astype('float')
```

```python
cols_keep=['County #',
'District #',
'2018-2019 ADA',
'2019-2020 ADA',
'2020-2021 ADA']

df_d[cols_keep].info()
```

```python
temp=[c + ': ' + str(df_d[c].isna().sum()) for c in df_d[cols_keep] if df_d[c].isna().any()]
temp
```

```python
df_d.to_csv('DATA_ADA_DISTRICT.csv', index = None)
```

```python
# read in the csv file
df_c = pd.read_csv('ada_county.csv')

df_c.drop(columns = [' 2004-2005 ADA', \
                    ' 2005-2006 ADA', \
                    ' 2006-2007 ADA', \
                    '2007-2008 ADA', \
                    '2008-2009 ADA', \
                    ' 2009-2010 ADA', \
                    '2010-2011 ADA', \
                    ' 2011-2012 ADA', \
                    ' 2012-2013 ADA', \
                    ' 2013-2014 ADA', \
                    ' 2014-2015 ADA', \
                    ' 2015-2016 ADA', \
                    ' 2016-2017 ADA'], inplace = True)

df_c
```
