***Data Reading:***

```python
# read in the csv file
df_d = pd.read_csv('ada data/ada_district.csv')

df_d
```

```python
df_d.drop(columns = [' 2004-2005 ADA', \
                      ' 2005-2006 ADA', \
                      ' 2006-2007 ADA', \
                      '2007-2008 ADA', \
                      '2008-2009 ADA', \
                      ' 2009-2010 ADA', \
                      '2010-2011 ADA', \
                      ' 2011-2012 ADA', \
                      ' 2012-2013 ADA', \
                      ' 2013-2014 ADA', \
                      ' 2014-2015 ADA', \
                      ' 2015-2016 ADA', \
                      ' 2016-2017 ADA'], inplace = True)

df_d.replace(',','', regex=True, inplace=True)
df_d['District #'] = df_d['District']
df_d[' 2017-2018 ADA']= df_d[' 2017-2018 ADA'].astype(float)
df_d[' 2018-2019 ADA']= df_d[' 2017-2018 ADA'].astype(float)
df_d[' 2019-2020 ADA']= df_d[' 2017-2018 ADA'].astype(float)
df_d[' 2020-2021 ADA']= df_d[' 2017-2018 ADA'].astype(float)
df_d.dropna(inplace = True)
zeros = df_d[df_d[' 2017-2018 ADA'] == 0]
df_d.drop(zeros.index, inplace=True)

df_d['2017-2018 ADA']= df_d[' 2017-2018 ADA']
df_d['2018-2019 ADA']= df_d[' 2018-2019 ADA']
df_d['2019-2020 ADA']= df_d[' 2019-2020 ADA']
df_d['2020-2021 ADA']= df_d[' 2020-2021 ADA']

df_d.drop(columns = [' 2017-2018 ADA', \
                     ' 2018-2019 ADA', \
                     ' 2019-2020 ADA', \
                     ' 2020-2021 ADA'], inplace = True)
df_d.info()
```

```python
df_d.to_csv('DATA_ADA_DISTRICT.csv', index = None)
```

```python
# read in the csv file
df_c = pd.read_csv('ada data/ada_county.csv')

df_c.drop(columns = [' 2004-2005 ADA', \
                    ' 2005-2006 ADA', \
                    ' 2006-2007 ADA', \
                    '2007-2008 ADA', \
                    '2008-2009 ADA', \
                    ' 2009-2010 ADA', \
                    '2010-2011 ADA', \
                    ' 2011-2012 ADA', \
                    ' 2012-2013 ADA', \
                    ' 2013-2014 ADA', \
                    ' 2014-2015 ADA', \
                    ' 2015-2016 ADA', \
                    ' 2016-2017 ADA'], inplace = True)

df_c
```

```python
columns = ['County', 'County Name', 'County #', 'Public Health Region', 'Health Service Region']

cc = pd.read_csv('ada data/county_conversion.csv', names = columns, header = 0)

cc.drop(columns = ['Public Health Region', 'Health Service Region', 'County Name'], inplace = True)

cc
```

```python
df_c = df_c.merge(cc, how = "inner", on = "County")
```
