````python
# column names
columns0 = ['District Name', \
          'District LEA #', \
          'District Enrollment 01/29/21', \
          'Approx. District On-Campus Enrollment 01/29/21', \
          'Campus Name', \
          'Campus ID', \
          'School Enrollment 01/29/21', \
          'School On-Campus Enrollment 01/29/21', \
          'New Student Cases', \
          'New Staff Cases', \
          'New School Reported Infection Source On-Campus', \
          'New School Reported Infection Source Off-Campus', \
          'New School Reported Infection Source Unknown', \
          'Cumulative Student Cases 08/01/21', \
          'Cumulative Staff Cases 08/01/21', \
          'Cumulative School Reported Infection Source On-Campus', \
          'Cumulative School Reported Infection Source Off-Campus', \
          'Cumulative School Reported Infection Source Unknown']

# read in the csv file
df0 = pd.read_csv('../covid data/covid_080121.csv', header = 6, names = columns0, na_values=["NR"])
df0.head(10)
````

````python
# drop new case columns and source columns   
df0.drop(columns = ['New Student Cases', \
                   'New Staff Cases', \
                   'New School Reported Infection Source On-Campus', \
                   'New School Reported Infection Source Off-Campus', \
                   'New School Reported Infection Source Unknown', \
                   'Cumulative School Reported Infection Source On-Campus', \
                   'Cumulative School Reported Infection Source Off-Campus', \
                   'Cumulative School Reported Infection Source Unknown'], \
                    inplace = True)

# extract 'Multiple Campus' instances, and remove them from the dataframe
#print(df0['Campus ID'].value_counts())
multiple_campus = df0[df0['Campus ID'] == 'Multiple\nCampus']
df0.drop(multiple_campus.index, inplace=True)
#print(df0['Campus ID'].value_counts())

# replace blank space with 0 or np.NaN
df0.replace(r'^\s*$', 0, regex=True, inplace=True)

# replace supressed data (*) with 3
df0.replace('*', 3, inplace = True)

# treat numeric features as float
df0['District Enrollment 01/29/21'] = df0['District Enrollment 01/29/21'].astype(float)
df0['Approx. District On-Campus Enrollment 01/29/21'] = df0['Approx. District On-Campus Enrollment 01/29/21'].astype(float)
df0['School Enrollment 01/29/21'] = df0['School Enrollment 01/29/21'].astype(float)
df0['School On-Campus Enrollment 01/29/21'] = df0['School On-Campus Enrollment 01/29/21'].astype(float)
df0['Cumulative Student Cases 08/01/21'] = df0['Cumulative Student Cases 08/01/21'].astype(float)
df0['Cumulative Staff Cases 08/01/21'] = df0['Cumulative Staff Cases 08/01/21'].astype(float)

# remove ' symbol from District ID
df0['District LEA #'] = df0['District LEA #'].astype(str)
df0['District LEA #'] = df0['District LEA #'].map(lambda x: x.lstrip("'"))

# remove schools that report to have no record of enrollment
df0.dropna(subset = ['School Enrollment 01/29/21'], inplace = True)

# create a new column for the % of on-campus students per district and per school
df0['% District On-Campus 01/29/21'] = df0['Approx. District On-Campus Enrollment 01/29/21'] / df0['District Enrollment 01/29/21']
df0['% School On-Campus 01/29/21'] = df0['School On-Campus Enrollment 01/29/21'] / df0['School Enrollment 01/29/21']

# create a new column for % of students (per school and per district) with COVID
# (note this is okay to calculate both as district total rows have District Enrollment = School Enrollment)
df0['% Students with COVID 08/01/21'] = df0['Cumulative Student Cases 08/01/21'] / df0['School Enrollment 01/29/21']
````

````python
# extract district totals and store them into a new dataframe
# then remove campus level information from the district dataframe
df0_d = df0[df0['Campus Name'].isna()]
df0_district = df0_d.drop(columns = ['Campus Name', 'Campus ID'])

df0_district.head(10)
````
